```python
from transformers import pipeline
```

```python
models = {
    "BERT": "bert-base-uncased",
    "RoBERTa": "roberta-base",
    "BART": "facebook/bart-base"
}
```

```python
prompt = "The future of Artificial Intelligence is"

for name, model in models.items():
    print(f"\nModel: {name}")
    try:
        generator = pipeline("text-generation", model=model)
        output = generator(prompt, max_length=30)
        print(output[0]["generated_text"])
    except Exception as e:
        print("Error:", e)
```

```python
sentence = "The goal of Generative AI is to [MASK] new content."

for name in ["BERT", "RoBERTa"]:
    print(f"\nModel: {name}")
    fill_mask = pipeline("fill-mask", model=models[name])
    outputs = fill_mask(sentence)
    for o in outputs[:3]:
        print(o["token_str"], "-", o["score"])
```

```python
sentence_bart = "The goal of Generative AI is to <mask> new content."

fill_mask_bart = pipeline("fill-mask", model=models["BART"])
outputs = fill_mask_bart(sentence_bart)

for o in outputs[:3]:
    print(o["token_str"], "-", o["score"])
```

```python
question = "What are the risks?"
context = "Generative AI poses significant risks such as hallucinations, bias, and deepfakes."

for name, model in models.items():
    print(f"\nModel: {name}")
    try:
        qa = pipeline("question-answering", model=model)
        answer = qa(question=question, context=context)
        print(answer)
    except Exception as e:
        print("Error:", e)
```

| Task | Model | Success / Failure | What I observed | Why it happened |
|------|-------|------------------|-----------------|-----------------|
| Text Generation | BERT | Failure | BERT could not generate proper text and showed errors. | BERT is an encoder-only model and not designed for text generation. |
| Text Generation | RoBERTa | Failure | RoBERTa also failed to generate meaningful text. | RoBERTa is encoder-only and lacks a decoder. |
| Text Generation | BART | Success | BART generated a meaningful continuation of the prompt. | BART is an encoder–decoder model designed for generation. |
| Fill Mask | BERT | Success | BERT correctly predicted words like create and generate. | BERT is trained using masked language modeling. |
| Fill Mask | RoBERTa | Success | It worked correctly after using `<mask>` instead of `[MASK]`. | RoBERTa uses a different mask token but is MLM-trained. |
| Fill Mask | BART | Partial | BART filled the mask but predictions were less accurate. | BART is not mainly trained for masked language modeling. |
| Question Answering | BERT | Partial | BERT gave an answer but it was not very accurate. | The model is not fine-tuned for question answering. |
| Question Answering | RoBERTa | Partial | RoBERTa produced weak or incomplete answers. | QA fine-tuning is required for better performance. |
| Question Answering | BART | Partial | BART returned partial answers with low confidence. | The QA head was not fine-tuned. |
